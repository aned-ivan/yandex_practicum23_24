##### Короткие и длинные поездки в городах
Можно предположить, что короткие поездки (до 1500 метров) чаще совершаются в городах с небольшой площадью, например, в Пятигорске (97 кв.м). А в городах с большой площадью территории, например в Москве ( коротких поездок меньше. 

Площадь городов, км²:
* Москва - 2 511
* Екатеринбург - 1 112
* Тюмень - 699 
* Омск - 573
* Ростов-на-Дону - 354
* Краснодар - 340
* Сочи - 177
* Пятигорск - 97

Добавим в датафрейм rides_go столбец с категорией поездки - короткая (до 1500 метров) и длинная (более 1500 метров), а также столбец с площадью города в датафрейм users_go. 

```python
# создадим формулу для категоризации расстояний
def categorize_distance(item):
    try:
        if 0 <= item <= 1500:
            return 'low_distance'
        elif item > 1500:
            return 'high_distance'
    except:
        pass
    
# создадим формулу для вставки площади городов 
def categorize_city(city_name):
    try:
        if city_name == 'Москва':
            return 2511
        elif city_name == 'Екатеринбург':
            return 1112
        elif city_name == 'Тюмень':
            return 699
        elif city_name == 'Омск':
            return 573
        elif city_name == 'Ростов-на-Дону':
            return 354
        elif city_name == 'Краснодар':
            return 340
        elif city_name == 'Сочи':
            return 177
        elif city_name == 'Пятигорск':
            return 97
    except:
        pass
```

```python
# Добавляем столбцы с категориями в датафрейм
rides_go['distance_category'] = rides_go['distance'].apply(categorize_distance)
users_go['city_area'] = users_go['city'].apply(categorize_city)

# создадим переменную и в ней для анализа объеденим датафреймы rides_go и users_go
rides_in_cities = users_go.merge(
    rides_go,
    on='user_id', # название столбца, по которому объединять, передают в параметре on
    how='inner', # режим объединения передают в параметре how
    suffixes=('_df_1', '_df_2')
)

rides_in_cities
```

```python
# посчитаем соотношение количество длинных поездок в каждом городе
high_rides_in_cities = rides_in_cities.query('distance_category == "high_distance"').pivot_table(index=['city', 'city_area'], values='distance_category', aggfunc='count').sort_values('distance_category', ascending=False).reset_index()

# посчитаем соотношение количество коротких поездок в каждом городе
low_rides_in_cities = rides_in_cities.query('distance_category == "low_distance"').pivot_table(index=['city', 'city_area'], values='distance_category', aggfunc='count').sort_values('distance_category', ascending=False).reset_index()

# переименуем названия столбцов
high_rides_in_cities.columns = ['city','city_area','count_high_distance']
low_rides_in_cities.columns = ['city','city_area','count_low_distance']

low_rides_in_cities
```

```python
high_rides_in_cities
```

```python
# объединим таблицы с количеством коротких и длинных поездок в городах
ratio_high_and_low_rides_in_cities = low_rides_in_cities.merge(
    high_rides_in_cities,
    on=['city', 'city_area'], # название столбца, по которому объединять, передают в параметре on
    how='inner', # режим объединения передают в параметре how
)

ratio_high_and_low_rides_in_cities
```

```python
# добавим столбцы с долей коротких и длинных поездок
# значения в них округлим до процентов
ratio_high_and_low_rides_in_cities['ratio_low_distance'] = \
    (round(ratio_high_and_low_rides_in_cities['count_low_distance']/(ratio_high_and_low_rides_in_cities['count_low_distance']+ratio_high_and_low_rides_in_cities['count_high_distance'])*100,2)) 

ratio_high_and_low_rides_in_cities['ratio_high_distance'] = \
    (round(ratio_high_and_low_rides_in_cities['count_high_distance']/(ratio_high_and_low_rides_in_cities['count_low_distance']+ratio_high_and_low_rides_in_cities['count_high_distance'])*100, 2)) 

#Отсортируем столбцы по проценту коротких поездок
ratio_high_and_low_rides_in_cities.sort_values('ratio_low_distance', ascending=False).reset_index()
```

По таблице видим, что разрыв в доле коротких и длинных поездок в городах с разной площадью составаляет всего 2,63%. Самая большая доля коротких поездок в Ростове-на-Дону (площадь города 354 кв.м.) — 11,35%.
Самая маленькая доля коротких поездок в Пятигорске (площадь города 97 кв.м.) — 8.72%.

Интересно, что самые разные по площади города - Москва (2511 кв.м.) и Пятигорск 97 (кв.м.) - стоят в рейтинге рядом и по доле коротких поездок, и по доле длинных.

То есть наше предположение о том, что площадь города влияет на долю коротких и длинных поездок - не подтвердилось.

```python
# Округлим продолжительность каждой до ближайшего числа вверх с попощью метода np.ceil() библиотеки numpy

all_data['duration'] = np.ceil(all_data['duration']).astype(int)

# Проверим, что данные о продолжительности округлились
all_data.head()
```

```python
# Добавим в столбец edited_all_data помесячную выручку, которую принёс каждый пользователь

# Помесячную выручку считаем по формуле: стоимость старта поездки × количество поездок + 
# + стоимость одной минуты поездки × общая продолжительность всех поездок в минутах +
# + стоимость подписки.

edited_all_data['monthly_revenue'] = \
    ((edited_all_data['start_ride_price'] * edited_all_data['amount_rides']) \
    +(edited_all_data['minute_price'] * edited_all_data['sum_duration']) + edited_all_data['subscription_fee'])

# напечатаем таблицу с добавленным столбцом выручки
edited_all_data
```

### Проверка гипотез
Продакт-менеджеры сервиса аренды самокатов GoFast хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции. Проверим для них несколько важных гипотез.

#### Подписчики тратят больше времени на поездки.
Если гипотеза верна, то пользователи с подпиской могут быть «выгоднее» для компании. 
Нулевая гипотеза: средняя продолжительность поездки подписчиков равна средней продолжительности поездки юзеров без подписки.

Альтернативная гипотеза: средняя продолжительность поездки подписчиков больше, чем средняя продолжительность поездки юзеров без подписки. То есть это правосторонняя гипотеза.

Проверим гипотезу, используем исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.

```python
# из исходного общего датафрейма создадим две выборки - время поездок подписчиков и неподписчиков

# подписчики
users_ultra = all_data.query('subscription_type == "ultra"')
users_ultra['duration']

# неподписчики
users_free = all_data.query('subscription_type == "free"')
users_free['duration']

# проверим, что обе выборки распределены нормально
users_free['duration'].hist(bins=50)
users_ultra['duration'].hist(bins=50)
plt.title('Продолжительность поездки')
plt.xlabel('Время, мин.')
plt.ylabel('Частота встречаемости')
plt.legend(['без подписки', 'подписчики']) # добавляем легенду
plt.show()
```

```python
print('Среднее время поездки юзеров без подписки:', round(users_free['duration'].mean(),2))

print('Среднее время поездки подписчиков:', round(users_ultra['duration'].mean(),2))
```

Обе выборки распределены нормально.

Средняя продолжительность поездок подписчиков больше, чем у юзеров без подписки. 
Чтобы точно принять или отвергнуть нулевую гипотезу, проведем двухвыборочный t-тест для независимых выборок.

```python
# задаём уровень статистической значимости в 5%
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем одностороннюю гипотезу «больше» (greater).
results = st.ttest_ind(users_ultra['duration'], users_free['duration'], alternative='greater')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве продолжительности поездок')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

###### Вывод
Гипотеза о равенстве продолжительности поездок отвергнута. 
Средняя продолжительность поездки подписчиков больше, чем средняя продолжительность поездки юзеров без подписки. С точки зрения времени поездки для компании выгоднее подписчики.

#### Помесячная выручка от подписчиков выше, чем от "неподписчиков"

Нулевая гипотеза: Помесячная выручка от пользователей с подпиской по месяцам равна выручке от пользователей без подписки.

Альтернативная гипотеза: Помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки. Правосторонняя гипотеза.

```python
# из исходного общего датафрейма создадим две выборки - выручка от подписчиков и неподписчиков

# подписчики
revenue_ultra = edited_all_data.query('subscription_type == "ultra"')
revenue_ultra['monthly_revenue']

# неподписчики
revenue_free = edited_all_data.query('subscription_type == "free"')
revenue_free['monthly_revenue']

# проверим, что обе выборки распределены нормально
revenue_free['monthly_revenue'].hist(bins=100)
revenue_ultra['monthly_revenue'].hist(bins=100)
plt.title('Выручка от пользователей')
plt.xlabel('Месячная выручка, руб.')
plt.ylabel('Частота встречаемости')
plt.legend(['без подписки', 'подписчики']) # добавляем легенду
plt.show()
```

```python
# сравним средние показатели выручки от подписчиков и юзеров без подписки

print('Средняя помесячная выручка без подписки:', round(revenue_free['monthly_revenue'].mean(),2))

print('Средняя помесячная выручка подписчиков:', round(revenue_ultra['monthly_revenue'].mean(),2))
```

Проверим гипотезу с помощью двухвыборочного t-тест для независимых выборок.

```python
# задаём уровень статистической значимости в 5%
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем одностороннюю гипотезу «больше» (greater).
results = st.ttest_ind(revenue_ultra['monthly_revenue'], revenue_free['monthly_revenue'], alternative='greater')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве выручки от подписчиков и юзеров без подписки')
else:
    print('Нет оснований нулевую гипотезу о равенстве выручки от подписчиков и юзеров без подписки')
```

#### Обновление сервера снизит количество обращений в техподдержку
Представим такую ситуацию: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Какой тест вам понадобился бы для проверки этой гипотезы?

#### Акция с промокодами
Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. 

Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 10% пользователей. Выясним, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5%. 

Подберем параметры распределения, описывающего эту ситуацию, построим график распределения и сформулируем ответ на вопрос о количестве промокодов.

```python
# найдем количество попыток продления подписки n - количество розданных промокодов.
# Подписку продлевает только каждый десятый клиент, для получения 100 подписок логично сделать 1000 попыток
n = 1000  

# по опыту прошлой акции вероятность продления подписки 10, укажем p = 0.1 
p = 0.1

distr = [] #создаем список distr, в котором будем хранить значения распределения

# построим распределение вероятностей
for k in range(0, n + 1):
    choose = factorial(n) / (factorial(k) * factorial(n-k))
    prob = choose * p**k * (1 - p)**(n - k) 
    distr.append(prob)
    
# построим гистограмму распределения вероятностей
plt.bar(range(0, n + 1), distr)
plt.show()
```

На гистограмме видно, что раздав 1000 промокодов с вероятностью продления подписки 0.1 у каждого промокода, наиболее вероятно получить примерно 100 продлений (количество смотрим по горизонтальной оси).

```python
# подберем точное количество провомодов с помощью перебора n
# мы перебором будем подходить к предельной вероятности ошибки 0.05 или 5%,
# постепенно увеличивая n на каждом шаге цикла

for n in range(700, 1500):
    if binom.cdf(99, n, p) <= 0.05:
        print("Необходимо", n, "промокодов")
        break
```

С помощью аппроксимации постройте примерный график распределения и оцените вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

Эта ситуация тоже описывается биномиальным распределением. Но считать каждое отдельное значение достаточно долго. Воспользуемся нормальной аппроксимацией биномиального распределения и методом cdf() для быстрой оценки.

```python
from math import sqrt

# задаём параметры биномиального распределения
n = 1_000_000
p = 0.4

# зададим мат.ожидание и ст.отклонение нормального распределения равными
# мат.ожиданию и ст.отклонению биномиального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение
distr = st.norm(mu, sigma) 

# считаем вероятность хвоста слева от 399,5 тысяч
result = distr.cdf(399500)
print(f'Вероятность, что уведомления откроют не более 399,5 тыс.пользователей равна: {result}')
```
