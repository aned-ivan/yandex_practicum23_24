## Введение

### Описание проекта
Интернет-магазин товаров для дома и быта «Пока все ещё тут» планирует рекламную рассылку по базе клиентов. Компания хочет сэкономить деньги на рассылке, сделав ее более адресной, чтобы рекламные предложения максимально соответствовали интересам и покупательскому поведению клиентов. 

Для осуществления адресной и качественной рассылки интернет-магазин просит сегментировать клиентов по профилю потребления на основе их истории покупок, а также дать подробные рекомендации по вззаимодействию с каждым выявленным сегментом.
В частности заказчика интересует:

1. В каком сегменте покупатели покупают часто, много и дорогой товар, а на какой сегмент, возможно, не стоит тратить время и деньги проводя рассылку,
2. Клиентам какого сегмента, когда и какие категории товара рассылать в рекламных предложениях,
3. Как и для какого кластера можно попытаться увеличить количество продаваемого товара, повысить возвращаемость покумателей.

###### Описание данных:
Для проведения исследования заказчик предоставил датафрейм, который описывает транзакции интернет-магазина товаров для дома и быта «Пока все ещё тут».

* date — дата заказа;
* customer_id — идентификатор покупателя;
* order_id — идентификатор заказа;
* product — наименование товара;
* quantity — количество товара в заказе;
* price — цена товара.

### Цели проекта: 
1. Сегментировать пользователей по профилю потребления для осуществления рекламных рассылок,
2. Составить подробные рекомендации заказчику по сегментам покупателей.

##### Сегментация пользователей методом RMF

##### Разбиение товаров по категориям (товары для кухни; растения; декор и др), создание столбца с названиями категорий

##### Подсчет среднего чека для каждого сегмента пользователей, визуализация динамики изменения среднего чека во времени

##### Подсчет и визуализация частоты покупок для каждого сегмента пользователей

##### Визуализация какие товары пользуются популярностью у каждой категории с течением времени

##### Подсчет и визуализация среднего количества товаров в заказе у каждой категории пользователей

##### Проверка на нормальность распределение среднего чека, среднего дней с последней покупки у каждой категории пользователей

##### Выбор статистического теста и сравнение кластеров пользователей для подтверждения, что пользователи верны разбиры на кластеры.

Каждый кластер будет сравнивать со всеми остальными кластерами.

Проверяться будут две пары нулевых и альтернативных гипотез:

1. Средний чек:
- H0 - Нет различий между средним чеком сегментов A и B. Сегменты клиентов не различаются, стоит пересмотреть разбивку на сегменты.
- H1 - Есть различия между средним чеком сегментов A и В. Сегменты клиентов различаются, сегментирование выполнено верно. 

2. Среднеее количество товаров в заказе:
- H0 - Нет различий между средним количествомтоваров в заказе у сегментов A и B. Сегменты клиентов не различаются, стоит пересмотреть разбивку на сегменты.
- H1 - Есть различия между средним количествомтоваров в заказе у сегментов A и B. Сегменты клиентов различаются, сегментирование выполнено верно.

##### Формулирование выводов 
1. В каком сегменте пользователи покумают часто, много и дорогой товар. 
2. Клиентам какого сегмента, когда и какие товары рассылать в рекламных предложениях. 
3. Как и для какого кламтера можно попытаться увеличить количество продаваемого товара, повысить возвращаемость покупателей.

#### Распределение заказов, выручки и ср. чека по месяцам

Добавим в датафрейм столбец year_month с годом и месяцев, чтобы посмотреть как менялись значения месяц от месяца.

```python
data['year_month'] = data['date_time'].dt.to_period('M')
data.head(5)
```

```python
# строим таблицу с количеством заказов и общей выручкой по месяцам
order_revenue_in_month = data.groupby('year_month').agg({'order_id': 'nunique', 'revenue': 'sum'})
order_revenue_in_month.columns = ['amount_order','sum_revenue']
order_revenue_in_month = order_revenue_in_month.reset_index()
order_revenue_in_month['avg_bill'] = round(order_revenue_in_month['sum_revenue']/order_revenue_in_month['amount_order'], 1)
order_revenue_in_month 
```

```python
# построим график с заказами
order_revenue_in_month.plot.bar(x='year_month', y='amount_order', grid=True, figsize=(10, 5))
plt.title('Общее количество заказов по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Заказы')
plt.legend('')
# Повернем годы на оси х под 45 градусов для удобства чтения
plt.xticks(rotation = 45) 
plt.show()
```

Наибольнее количество заказов интернет-магазин собрал в феврале 2019 года - 289 шт. В мае-сентябре 2019 года количество заказов заметно просело (до 184) и начало стабильно расти в октябре 2019 года. В январе 2020 года количество заказов достигло 269.  

```python
# построим график выручки
order_revenue_in_month.plot.bar(x='year_month', y='sum_revenue', grid=True, figsize=(10, 5))
plt.title('Суммарная выручка по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Выручка')
plt.legend('')
# Повернем годы на оси х под 45 градусов для удобства чтения
plt.xticks(rotation = 45) 
plt.show()
```

Суммарная выручка имеет тенденцию к снижению - с 351 910 руб в октябре 2018г до 174 028 руб в январе 2020 года.

```python
# построим график выручки
order_revenue_in_month.plot.bar(x='year_month', y='avg_bill', grid=True, figsize=(10, 5))
plt.title('Средний чек на заказ по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Ср.чек')
plt.legend('')
# Повернем годы на оси х под 45 градусов для удобства чтения
plt.xticks(rotation = 45) 
plt.show()
```

### Сегментация пользователей методом RMF
Сегментировать покупателей будем по трем характеристикам их поведения. 
* Recency (Давность) — Как давно клиент совершил покупку? Количество дней с момента последней покупки.
* Frequency (Частота) — Как часто они совершают покупки? Общее количество заказов.
* Monetary Value (Денежная ценность) — Сколько они тратят? Средний чек клиента в интернет магазине.

Дата последнего заказа в нашем датафрейме — 2020-01-31. Используем ее для расчета давности каждого клиента, т.е. количества дней с момента его последней покупки.

```python
rfm_data = data[['customer_id', 'order_id', 'quantity', 'date_time', 'revenue']]
rfm_data
```

Для дальнейшей сегментации используем метод qcut.

```python
NOW = dt.datetime(2020,1,31)

# Создаем таблицу с показаделями давности, часторы и тран для каждого уникального клиента
rfm_table=rfm_data.groupby('customer_id').agg({'date_time': lambda x: (NOW - x.max()).days, # Вычисляем давность (Recency)
                                                'order_id': lambda x: len(x.unique()), # Вычисляем частоту (Frequency)
                                                'revenue': lambda x: x.sum()})    # Вычисляем сумму трат (Monetary) 

rfm_table['date_time'] = rfm_table['date_time'].astype(int)
rfm_table['order_id'] = rfm_table['order_id'].astype(int)

# считаем средний чек для каждого клиента
rfm_table['monetary_value'] = rfm_table['revenue']/rfm_table['order_id']
rfm_table['monetary_value'] = rfm_table['monetary_value'].astype(int)

# переименуем столбцы
rfm_table.rename(columns={'date_time': 'recency', 
                         'order_id': 'frequency',
                         'monetary_value':'monetary_value'}, inplace=True)

# удалим колонку revenue, чтобы не путаться
del rfm_table['revenue']

rfm_table.head(5)
```

```python
# # задаем ярлыки групп
r_labels = range(1,5); f_labels = range(1, 5); m_labels = range(1, 5)

# присваиваем эти ярлыки 4 равным группам методом qcut
r_groups = pd.qcut(rfm_table['recency'], 4, labels = r_labels)

m_groups = pd.qcut(rfm_table['monetary_value'], 4, labels = m_labels)

rfm_table = rfm_table.assign(r=r_groups.values, m=m_groups.values)
rfm_table.head(5)
```

Для характеристики frequency сделаем разбивку вручную, т.к.в датасете очень много единичных заказов — и метод qcut не может разбить их на равные группы. 

Cделаем разбивку 1 заказ - группа 1, 2 заказа - группа 1, 3 заказа - группа 3, 4 и более заказов - группа 4.

```python
# Напишем функцию для разбивки. Аргументы (x = value, p = recency, monetary_value, frequency, d = quartiles dict)
def F_Class(x):
    if x == 1:
        return 1
    if x == 2:
        return 2
    if x == 3:
        return 3
    return 4
```

```python
rfm_table['f'] = rfm_table['frequency'].apply(F_Class)

rfm_table
```

```python
print('Ранги и количество клиентов по давности заказов:')
rfm_table.groupby('r')['recency'].agg(['mean','count']).rename(columns={'mean':'Средняя давность, дней','count':'Количество клиентов, чел'}).sort_values(by='r', ascending=False)
```

```python
print('Ранги и количество клиентов по частоте заказов:')
rfm_table.groupby('f')['frequency'].agg(['mean','count']).rename(columns={'mean':'Средняя частота заказов, шт','count':'Количество клиентов, чел'}).sort_values(by='f', ascending=False)
```

```python
print('Ранги и количество клиентов тратам:')
rfm_table.groupby('m')['monetary_value'].agg(['mean','count']).rename(columns={'mean':'Средний чек, руб','count':'Количество клиентов, чел'}).sort_values(by='m', ascending=False)
```

Далее посчитаем общий балл каждого клиента rfm_score, который будет характеризовать лояльность клиентов. Суммируем r, f, m и разделим клиентов на более крупные кластеры Lost Customer (потеряные клиенты), Low-value customer (клиенты с низкой ценностью), Medium value customer (клиенты со средней ценностью), Top Customer (лучшие клиенты).

```python
rfm_table['rfm_score'] = rfm_table['r'].astype(np.int8) + rfm_table['f'].astype(np.int8) + rfm_table['m'].astype(np.int8)
rfm_table = rfm_table.sort_values(by='rfm_score', ascending=False).reset_index()
rfm_table
```

Сгруппируем клиентов по общему баллу rfm_score.

```python
rfm_score_rating = rfm_table.groupby('rfm_score').agg({'recency': 'mean', 'frequency': 'mean', 'monetary_value': 'mean', 'customer_id': 'nunique'}).round(2).sort_values(by='rfm_score', ascending=False).reset_index()
rfm_score_rating = rfm_score_rating.rename(columns={'recency':'mean_recency','frequency':'mean_frequency','monetary_value':'mean_monetary','customer_id':'quantity_clients' })
rfm_score_rating
```

По таблице видно, что наибольший общий балл (12-9) имеют клиенты, которые делали покупки около года назад (в среднем от 352 до 463 дней). они совершали от 3 до 35 покупок (35 покупок сделал всего 1 клиент). В этой группе общей сложности это 215 клиентов. Средний чек у них высокий - от 1636 до  5435 руб.

Объединим клиентов в сегменты. Покупателей с общим баллом rfm_score от 12 до 9 объединим в группу Top Customer, с баллами от 8 до 7 - в группу Medium value customer,  с баллами от 6 до 5 - Low-value customer, с баллами от 4 до 3 - Minimal-value Customer.

```python
def categorize_rfm_segment(rfm_score):
    if rfm_score >= 9:
        return 'Top customer'
    elif (rfm_score == 8) or (rfm_score == 7):
        return 'Medium-value customer'
    elif (rfm_score == 6) or (rfm_score == 5):
        return 'Low-value customer'
    elif rfm_score <= 4:
        return 'Minimal-value customer'
```

```python
rfm_table['rfm_segment'] = rfm_table['rfm_score'].apply(categorize_rfm_segment)
rfm_table
```

#### Проверка сегментов стат. методом.
Проверим выявленные сегменты с помощью статистического теста. Если тест найдет статистически значимую разницу в сравниваемых сегментах, значит разбиение на сегменты выполнено верно.
Проверять будем отличия по среднему чеку (monetary_value) и давности (recency).

###### Гипотезы по среднему чеку:

* H0 - Нет различий между средним чеком сегментов A и B. Сегменты клиентов не различаются, стоит пересмотреть разбивку на сегменты.
* H1 - Есть различия между средним чеком сегментов A и В. Сегменты клиентов различаются, сегментирование выполнено верно.

###### Гипотезы по давности:
* H0 - Нет различий между средним количеством дней с последнего заказа у сегментов A и B. Сегменты клиентов не различаются, стоит пересмотреть разбивку на сегменты.
* H1 - Есть различия между средним количеством дней с последнего заказа у сегментов A и B. Сегменты клиентов различаются, сегментирование выполнено верно.

Применим поправку Бонферрони, потому что делаем множественные сравнениях групп.

```python
def compare_segments(data, alpha, n):
    
    """Эта функция сравнивает уникальные сегменты из колонки RFMClass по значениям колонок monetary_value и recency.

    bonferroni_alpha = alpha / n — критический уровень статистической значимости c поправкой Бонферрони
     
    """
    alpha = alpha
    bonferroni_alpha = alpha / n
    segments = data['rfm_segment'].unique()
    
    significant_monetary_diff = []
    non_significant_monetary_diff = []
    significant_recency_diff = []
    non_significant_recency_diff = []
    
    for segment1 in segments:
        for segment2 in segments:
            if segment1 != segment2:
                data1_monetary = data[data['rfm_segment'] == segment1]['monetary_value']
                data2_monetary = data[data['rfm_segment'] == segment2]['monetary_value']
                
                results_monetary = st.mannwhitneyu(data1_monetary, data2_monetary)
                
                print(f"Сравнение сегментов {segment1} и {segment2} по monetary_value:")
                print('p-value:', results_monetary.pvalue)
                
                if results_monetary.pvalue < bonferroni_alpha:
                    print("Отвергаем нулевую гипотезу: разница статистически значима")
                    significant_monetary_diff.append((segment1, segment2))
                else:
                    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
                    non_significant_monetary_diff.append((segment1, segment2))
                    
                data1_recency = data[data['rfm_segment'] == segment1]['recency']
                data2_recency = data[data['rfm_segment'] == segment2]['recency']
                
                results_recency = st.mannwhitneyu(data1_recency, data2_recency)
                
                print(f"Сравнение сегментов {segment1} и {segment2} по recency:")
                print('p-value:', results_recency.pvalue)
                
                if results_recency.pvalue < bonferroni_alpha:
                    print("Отвергаем нулевую гипотезу: разница статистически значима")
                    significant_recency_diff.append((segment1, segment2))
                else:
                    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
                    non_significant_recency_diff.append((segment1, segment2))
                
                print('\n')
    return significant_monetary_diff, non_significant_monetary_diff, significant_recency_diff, non_significant_recency_diff

# Вызываем функцию для сравнения сегментов
significant_monetary, non_significant_monetary, significant_recency, non_significant_recency = compare_segments(united_data, 0.05, 12)
```

```python
def categorize_goods(goods_name):
    if goods_name in electronics:
        return 'Электроника'
    elif goods_name in hallway:
        return 'Интерьер'
    elif goods_name in plants:
        return 'Растения'
    elif goods_name in cleaning:
        return 'Чистящие средства'
    elif goods_name in kitchen:
        return 'Кухня'
    elif goods_name in coatings:
        return 'Текстиль и ковры'
    elif goods_name in bags:
        return 'Сумки и тележки'
    elif goods_name in wc:
        return 'Санузел'
    elif goods_name in household:
        return 'Хозтовары'
```

```python
quantity_goods_in_category = united_data.groupby('goods_category').agg({'product': 'nunique', 'price': 'mean', 'order_id': 'nunique', 'customer_id': 'nunique', 'revenue': 'sum'}).round(2).sort_values(by='product', ascending=False).reset_index()
quantity_goods_in_category = quantity_goods_in_category.rename(columns={'product': 'quantity_product', 'price': 'mean_price', 'order_id': 'quantity_orders', 'customer_id': 'quantity_clients', 'revenue': 'revenue'})

# доли клиентов, выручки и заказов для каждой категории товаров 
quantity_goods_in_category['share_of_clients, %'] = round(quantity_goods_in_category['quantity_clients']/quantity_goods_in_category['quantity_clients'].sum()*100,2)
quantity_goods_in_category['share_of_revenue, %'] = round(quantity_goods_in_category['revenue']/quantity_goods_in_category['revenue'].sum()*100,2)
quantity_goods_in_category['share_of_orders, %'] = round(quantity_goods_in_category['quantity_orders']/quantity_goods_in_category['quantity_orders'].sum()*100,2)
quantity_goods_in_category['share_of_quantity, %'] = round(quantity_goods_in_category['quantity_product']/quantity_goods_in_category['quantity_product'].sum()*100,2)
quantity_goods_in_category
```

######  RFM-анализ

В ходе исследования потребительского поведения проведен RFM-анализ и выявлены сегменты клиентов отличающиеся по сроку с последней покупки, частоте заказов и среднему чеку на заказ. Сегментация проверена с помощью статистического теста Манна-Уитни. Сегменты, в которых тест не обнаружил статистически значимой разницы, объединены.

###### Сформированы следующие группы сегментов и рекомендации для этих клиентов:

Сформированы 4 сегмента клиентов по их покупательскому поведению.

###### Сегмент Top customer (лучшие клиенты)
Сегмент Top customer (лучшие клиенты) - 215 клиентов (8,91% от всех покупателей), принесли 28,78% общей выручки магазина за исследуемый период и 12,72% заказов. Клиенты этого сегмента имеют самый высокий средний чек (2867,88 руб) и самое высокое количество заказов (23.75 шт). В то же время этот сегмент имеет самую большую давность с последней покупки. Самые лучшие и платящие клиенты не заходят в интернет магазин в среднем 408 дней. Нужно попытаться их вернуть с помощью рекламной рассылки.

###### Сегмент Medium-value customer 
Сегмент Medium-value customer (клиенты средней ценности) - 904 человека(41,25% от всех покупателей), принесли 50.31% общей выручки магазина за исследуемый период и 33.37% заказов. Средняя давность с последней покупки около 265 дней - меньгше чем у первой группы, но все равно высокий показатель. Для них нужна рекламная расылка, чтобы вернуть их к покупкам. Для них рекоменндуем формулировать предложения для увеличения частоты покупок. Готовность тратить (чек) у них достаточно хорошая, с увеличением частоты покупок эти клиенты могут принести значительную выгоду.

Для них тоже нужна рекламная рассылка, но акцент предлагаем сделать на увеличении среднего чека. Это многочисленная категория, которая делает 40% заказов, но приобретает недорогие товары.
