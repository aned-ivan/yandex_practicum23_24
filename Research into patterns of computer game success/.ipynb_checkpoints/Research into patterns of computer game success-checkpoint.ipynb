# Исследование закономерностей успешности компьютерных игр

## Введение

### Описание проекта

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников нам доступны исторические данные  (до 2016 года) о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

### Цель проекта

Провести исследовательский и статисчтический анализ данных, выявить закономерности, которые деляют компьютерную игру успешной.

### Описание данных

Полученный для исследования датафрейм имеет следующую информацию: 

* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Важно: Данные за 2016 год могут быть неполными.

Путь к файлу: /datasets/games.csv. 

ССылка для скачивания: https://code.s3.yandex.net/datasets/games.csv

В нашей таблице 8 возрастных категорий. Расшифруем их:

* «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.


* «E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов (ранее с 6 лет). Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально "K-A" ("Kids to Adults"). То есть категории «E»  и «К-А» — одно и тоже.


* «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.


* «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.


* «M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.


* «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.


* «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

Указать вместо пропусков точный год выпуска игры нет проблематично. Игры могли выходить в разное время в зависимости от стран и платформ - какой именно год указывался в year_of_release можно только предполагать.

Восстановить оценку критиков невозможно. Заменять пропуски медиаными оценками, например в зависимости от жанра или года выпуска, будет неправильным, т.к. внутри жанра или года игры могут иметь очень разное качество. 
Заменим пропуски загрушками: - 1.

```python
# заменим пропуски на -1
games['critic_score'] = games['critic_score'].fillna('-1')

print('Количество пропущенных значений в колонке genre:', games.loc[games['critic_score'].isna()]['critic_score'].count())

# проверим, что пропуски заполнены заглушкой 
games.loc[games['critic_score'] == "-1"]
```

```python
 # смотрим уникальные значения в колонке user_score
games['user_score'].unique().tolist()
```

В колонке user_score обнаружено аномальное значение 'tbd' - "подлежит уточнению". Уточненную оценку мы ее можем восстановить, поэтому заменим значения 'tbd' на заглушку -1.
Пропущенные оценки пользователей также заменим на заглушку -1, т.к. восстановить точныке оценки невозможно.

```python
# меняем 'tbd' на '-1'
games = games.replace('tbd', '-1')

# заменим пропуски на -1
games['user_score'] = games['user_score'].fillna('-1')

# проверим, что пропуски заполнены заглушкой
print('Количество пропущенных значений в колонке user_score:', games.loc[games['user_score'].isna()]['user_score'].count())
games['user_score'].unique().tolist()
```

```python
# Рассмотрим пропуски в rating
games.loc[games['rating'].isna()]
```

Рейтинг ESRB был запущен в 1994 году для контента выпускаемого в США и Канаде. Поэтому для игр продававшихся в Северной Америке до 1994 года значения рейтинга отсутствуют. Плюс данный рейтинг использовался только для С.Америки. В других регионах игры могли издаваться без рейтинга ESRB, либо с рейтингами других стран (например, CERO в Японии с 2002 года, или PEGI в Европе с 2003 года).  

Удалять пропуски - плохое решение, т.к. пропусков очеть много - это повлияет на результат исследования. Восстановить пропущенные рейтинги проблематично. Можно попробовать спарсить данные с сайта ESRB или другого рейтингового агентства, но в рамках данного исследования ограниченного жестким дедлайном, нам удобнее заменить пропуски заглушкой 'unknown rating'.

Типы данных логичны не во всех столбцах.

Данные в колонках name, platform, genre, rating содержат текст. Лучше переделать тип на str.
Колонка year_of_release содержит год (целое число, а не дробь), поэтому лучше переделать тип на int.

В колонках na_sales, eu_sales, jp_sales, other_sales оставим имеющийся тип float64, т.к. количество проданных копий (млн.штук) может быть не круглым.

#### Влияние отзывов на продажи

Посмотрим, как влияют на продажи внутри платформ отзывы пользователей и критиков. Для примера рассмотрим платформы с суммарными продажами более 50 млн шт. в 2012-2016 годах. 

```python
# составляем список платформ
list_platforms = sales_in_recent_years.query('sum_total_sales >= 50')['platform'].unique().tolist()
list_platforms 
```

##### Отзывы критиков
Сначала построим диаграмму рассеяния для оценок критиков и посчитаем корреляцию между отзывами и продажами.

```python
# отключим вывод предупреждений на экран
import warnings
warnings.filterwarnings("ignore") 

# сохраняем в переменную строки с известными оценками критиков
clear_critic_score = new_data_games.query('critic_score != -1')

# построиm диаграммы распределения с помощью цикла
for platform in list_platforms:
    data = clear_critic_score.query('platform == @platform')[['total_sales', 'critic_score']]
    
    fig, ax = plt.subplots(figsize = (7,7))
    pd.plotting.scatter_matrix(data, ax=ax)
    fig.suptitle(f'Корреляция продаж с оценками критиков на платформе {platform}')
    plt.show()   
    
    print(f'Кореляция Пирсона {platform}:')
    display(clear_critic_score.query('platform == @platform')[['total_sales','critic_score']].agg('corr'))
```

```python
# Комментарий ревьюера
print(len(new_data_games.loc[new_data_games['critic_score'] != -1]))
print(len(new_data_games.loc[new_data_games['user_score'] != -1]))
print(len(new_data_games.loc[(new_data_games['critic_score'] != -1)&(new_data_games['user_score'] != -1)]))
```

###### Вывод
Диаграммы распределений топ-7 платформ по суммарным продажам показывают небольшую положительную корреляцию (от 0,23 до 0,41) между оценками критиков и колличеством проданных копий игры. То есть чем выше оченка критиков, тем больше продаж игр соответствующей платформы. 
Можно сказать, что хорошие отзывы в состоянии немного увеличить продажи игр, но радикально повлиять на продажи не могут.

##### Отзывы пользователей
По аналогии построим диаграммы рассеяния для оценок пользователей и посчитаем корреляцию между их отзывами и продажами игр.

```python
# сохраняем в переменную строки с известными оценками критиков
clear_user_score = new_data_games.query('user_score != -1')

# построиm диаграммы распределения с помощью цикла
for platform in list_platforms:
    data = clear_user_score.query('platform == @platform')[['total_sales', 'user_score']]
    
    fig, ax = plt.subplots(figsize = (7,7))
    pd.plotting.scatter_matrix(data, ax=ax)
    fig.suptitle(f'Корреляция продаж с оценками пользователей на платформе {platform}')
    plt.show()   
    
    print(f'Кореляция Пирсона {platform}:')
    display(clear_user_score.query('platform == @platform')[['total_sales','user_score']].agg('corr'))
```

###### Вывод
Между оценками пользователей и количеством проданных игр мы видим разброс корреляции.
На двух платформах мы видим небольшую положительную корреляцию — 3ds (0,19) и wiiu (0,40).
На остальных пяти платформах — отсутствие корреляции или небольшую отрицательную корреляцию: от -0,12 (рс) до 0.006 (х360).

Резюмируя можно сказать, что на продажи игр на платформах больше влияют оценки критиков. Оценки пользователей виеют влияние на продажи лишь в отдельных случаях, на отдельных платформах.
Оценки критиков влияют на продажи сильнее, чем оценки пользователей, вероятно потому, что критики играют в новинки и оценивают их первыми, до массового запуска в продажу (компании стараются предоставить свой продукт на пробу лидерам мнений на этапе запуска для его раскрутки и рекламы). Пользователи ориентируются на отзывы критиков, а свои оценки выставляют уже после покупки игры (уже обеспечив своими деньгами часть продаж игры).

#### Влияние рейтинга на продажи
Изучим влияет ли рейтинг ESRB на продажи в отдельном регионе.

```python
# Добавим столбец с описанием категориии рейтинга 
# создадим формулу для вставки площади городов 
def categorize_rating(rating_name):
    try:
        if rating_name == 'e':
            return 'Для детей младшего возраста, от 3 лет'
        elif rating_name == 'ec':
            return 'Для всех, от 6 лет'
        elif rating_name == 'e10+':
            return 'Для всех, от 10 лет'
        elif rating_name == 'm':
            return 'Для взрослых, от 17 лет'
        elif rating_name == 't':
            return 'Подросткам, от 13 лет'
        elif rating_name == 'ao':
            return 'Только для взрослых, 18+'
        elif rating_name == 'rp':
            return 'Рейтинг ожидается'
        elif rating_name == 'unknown rating':
            return 'Нет рейтинга'
    except:
        pass   
```

```python
# Добавляем столбцы с категориями в датафрейм
new_data_games['rating_category'] = new_data_games['rating'].apply(categorize_rating)

new_data_games
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
# из исходного общего датафрейма создадим две выборки - пользовательские рейтинги платформ Xbox One и PC

# юзерыXbox One
user_score_XOne = new_data_games.loc[(new_data_games['user_score'] != -1)&(new_data_games['platform'] == 'xone')]
user_score_XOne['user_score']

# юзеры PC
user_score_PC = new_data_games.loc[(new_data_games['user_score'] != -1)&(new_data_games['platform'] == 'pc')]
user_score_PC['user_score']

# проверим, что обе выборки распределены нормально
user_score_XOne['user_score'].hist(bins=50)
user_score_PC['user_score'].hist(bins=50)
plt.title('Пользовательские рейтинги платформ Xbox One и PC')
plt.xlabel('Рейтинг пользователей.')
plt.ylabel('Частота встречаемости')
plt.legend(['XOne', 'PC']) # добавляем легенду
plt.show()
```

Обе выборки распределены нормально. Распределение оценок юзеров PC немного смещены вправо относительно оценок юзеров XOne.

Проведем двухвыборочный t-тест для независимых выборок. 

Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC равны.

Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC не равны.

Из общепринятых уровней статистической значимости (alpha) 1% и 5% выберем 5%. Точности в 95% будет достаточно для нашего исследования. 

```python
# задаём уровень статистической значимости в 5%, 
alpha = 0.05

# проводим ttest для двух выборок, аргумент alternative не указывае, чтобы проверилась двухсторонняя гипотеза
# укажем equal_var=False, т.к. нет уверенности, что дисперсии генеральных совокупностей равны
results = st.ttest_ind(user_score_XOne['user_score'], user_score_PC['user_score'], equal_var=False)

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних оценок пользователей Xbox One и PC')
else:
    print('Нет оснований отвергнуть нулевую гипотезу о равенстве средних оценок пользователей Xbox One и PC')
```

```python
print('Средняя оценка юзеров XOne:', round(user_score_XOne['user_score'].mean(),2))

print('Средняя оценка юзеров PC:', round(user_score_PC['user_score'].mean(),2))
```

###### Вывод
Разница в средней оценке юзеров XOne и РС всего 0,11. Статистический тест показывает, что мы не может отвергнуть нулевую гипотезу. Значит принимаем ее - средние пользовательские рейтинги платформ Xbox One и PC равны.   

#### Средние пользовательские рейтинги жанров Action и Sports разные

```python
# из исходного общего датафрейма создадим две выборки - пользовательские рейтинги жанров Action и Sports

# жанр Action
user_score_action = new_data_games.loc[(new_data_games['user_score'] != -1)&(new_data_games['genre'] == 'action')]
user_score_action['user_score']

# жанр Sports
user_score_sports = new_data_games.loc[(new_data_games['user_score'] != -1)&(new_data_games['genre'] == 'sports')]
user_score_sports['user_score']

# проверим, что обе выборки распределены нормально
user_score_action['user_score'].hist(bins=50)
user_score_sports['user_score'].hist(bins=50)
plt.title('Пользовательские рейтинги жанров Action и Sports')
plt.xlabel('Рейтинг пользователей.')
plt.ylabel('Частота встречаемости')
plt.legend(['Action', 'Sports']) # добавляем легенду
plt.show()
```

Обе выборки распределены нормально. Наиболее частотные оценки жанров Action немного смещены вправо относительно оценок Sports.

Проведем двухвыборочный t-тест для независимых выборок.

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные.

Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные не равны.

Из общепринятых уровней статистической значимости (alpha) 1% и 5% выберем 5%. Точности в 95% будет достаточно для нашего исследования.

Аргумент alternative не будем указывать, тк нам нужно проверить отличаются ли выборки в какую-нибудь сторону.

```python
# задаём уровень статистической значимости в 5%, 
alpha = 0.05

# проводим ttest для двух выборок, аргумент alternative не указывае, чтобы проверилась двухсторонняя гипотеза
# укажем equal_var=False, т.к. нет уверенности, что дисперсии генеральных совокупностей равны
results = st.ttest_ind(user_score_action['user_score'], user_score_sports['user_score'], equal_var=False)

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних оценок жанров Action и Sports')
else:
    print('Нет оснований отвергнуть нулевую гипотезу о равенстве средних оценок жанров Action и Sports')
```

```python
print('Средняя оценка жанра Action:', round(user_score_action['user_score'].mean(),2))

print('Средняя оценка жанра Sports:', round(user_score_sports['user_score'].mean(),2))
```

###### Вывод
На имеющихся данных и на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы. Разница в средней оценке жанров Action и Sports =  1,37.

Учитывая, что японские пользователи предпочитают японские игровые платформы, то логично предположить, что игры без рейтинга (без рейтинга ERBS) - это японские игры изданные для местного рынка.

* Игры без рейтинга в Америке и Европе занимают третью по счету долю рынка (19-22%). Японские игры вполне могут занимать такую долю на рынках этих регионов.

* Игры "Для взрослых, от 17 лет", которые лидируют в Америке и Европе с почти 40% рынка, в Японии занимают только 4 место (11% рынка).	


###### Для С.Америки рекомендуем разрабатывать игры для взрослой аудитории в жанрах Action, Shooter, Sports, Role-Playing, Misc. Выпускать их в первую очередь на растущей и привычной американцам платформе Xоnе, во вторую очередь на растущей платформе PS4. 

###### Для Европейского рынка нужны игры для взрослой аудитории в жанрах Action, Shooter, Sports, Role-Playing, Racing. Выпускать их в первую очередь на растущей платформе PS4, во вторую очередь на Xоnе.

###### Чтобы быть успешными на японском рынке, игра должна ориентироваться в первую очередь на портативную платформу 3DS (это региональная особенность). Во-вторую, очередь обратить внимание на растущую платформу PS4. Игры должны быть в жанрах: Role-Playing, Action, Misc, Simulation Fighting. Предпочтительные возрастные категории для японского рынка необходимо изучать дополнительно, так лидирующей возрастной катгорией может оказаться как 18+, так и 6+.
